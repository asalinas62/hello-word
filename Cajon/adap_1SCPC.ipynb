## Adaptación mediante la línea

```python
from lines import *
```

```python
zln=0.5+0.3j
```

Primeramente hay que localizar el punto en la carta y trazar su círculo de $\Gamma$.

El coeficiente de reflexión en la carga tiene un módulo de 0.632 y una fase de 108 grados. En componentes cartesianas (parte real e imaginaria) tendríamos $\Gamma_L=-0.2+\jmath 0.6$. La representación en la carta será:

```python
ga=Gr(zln)
gam=abs(ga)
gaf=np.angle(ga)
fig=plt.figure()
fig.set_size_inches(5.7,5.7)
Skele()
Point(ga.real,ga.imag,'y')
Circle(gam,'y')
Cirx(zln.imag,c='r')
Cirr(zln.real,c='b')
#linea(0,0,*P2c(1.1,np.angle(ga)),'y')
#arco(1.1,0,np.angle(ga),'y')
plt.show()
```

El siguiente paso es determinar la distancia que hay desde la carga hasta el punto en el que la impedancia normalizada de la línea tiene como parte real la unidad. Se trata de encontrar las intersecciones del circulo de $\Gamma$ (amarillo) con el círculo $r=1$ (azul).

```python
fig=plt.figure()
fig.set_size_inches(5.7,5.7)
Skele()
Point(ga.real,ga.imag,'y')
Circle(gam,'y')
#Cirx(zln.imag,c='r')
Cirr(1,c='b')
#linea(0,0,*P2c(1.1,np.angle(ga)),'y')
#arco(1.1,0,np.angle(ga),'y')
plt.show()
```

```python
zpn1_uno,zpn2_uno=(0.09740719749811459,0.28516588859579)
x1_uno=Zwn(zpn1_uno,zln).imag
x2_uno=-x1_uno
dn1_uno=0.3902507992050169
dn2_uno=0.5-dn1_uno
gaf_es=(pi-gaf)/(4.*pi)
zpn1_uno_rad=zpn1_uno*4.*pi
zpn2_uno_rad=zpn2_uno*4.*pi
dn1_uno_rad=dn1_uno*4.*pi
dn2_uno_rad=dn2_uno*4.*pi
```

```python
fig=plt.figure()
fig.set_size_inches(6,6)
Skele()
Point(ga.real,ga.imag,'y')
Circle(gam,'y')
Line(0,0,*P2c(1.1,gaf),'y')
Cirr(1,'b')
Point(*P2c(gam,gaf-zpn1_uno_rad),'g')  # Punto verde
Line(0,0,*P2c(1.1,gaf-zpn1_uno_rad),'g')  # Linea verde
Arc(1.1,gaf,gaf-zpn1_uno_rad,'g')  # Arco verde
Point(*P2c(gam,gaf-zpn2_uno_rad),'m')  # Punto magenta
Line(0,0,*P2c(1.1,gaf-zpn2_uno_rad),'m')  # Linea magenta
Arc(1.05,gaf,gaf-zpn2_uno_rad,'m')
plt.text(0.35,0.4,'1')
plt.text(0.35,-0.4,'2')
#Cirx(xruno1,'r')
#punto(*P2c(np.abs(ga),np.angle(ga)-zn2*(4*pi)),'r')
#cirx(xruno2,'r')
plt.show()
```
